# 1. Introduction to the O&G industry
Okay so that put aside, lets get into the main event, lets cover a little bit of background about Oil and Gas for the new comers: 
Oil and gas is located in porous deposits which are often located thousands of feet below the earths surface. The process of extracting oil and gas from these depths to the surface, and then converting it into a multi-usable source involves a large variety of operations. Broadly speaking, the entire process of extracting value from Oil and Gas can be divided into the following three industry sectors. Up, mid, and Downstream

Each one of them including a number of operations and each with its own science quite honestly, but they are highly correlated, meaning that they heavily depend on each other.


## 1.1 Upstream Sector
### 1.1.1 Exploration and Appraisal
finding the exact location of oil and gas deposits under the earths surface is an activity as challenging as finding a needle in the haystack. Achieving that is the role of the subsurface people, mainly geologists and geophysicists   
Geological surveys provide geological maps or models of the structure under the earths surface.
While geophysical assessment using various technics such as magnetic, electromagnetic, gravimetric, and seismic surveys try to quantify or predict the nature of subsurface using the own rock behaviour.

Once potential locations of the petroleum reserves are identified with a certain level of confidence, the operating oil and gas company acquires a lease of the oil and gas block. A wildcat is drilled to verify potential oil and gas reserve locations. If oil is found in the first wildcat well, reservoir engineers get involved in the field appraisal process. 

### 1.1.2 Drilling and Completion
The process of drilling focuses on creating pathways for oil and gas to flow from the oil and gas reservoirs to the earths surface. Drilling is a high risk and high capital expenditure (CAPEX) operation. As a ballpark figure, a single offshore well may cost from 100 million US dollars to 200 millionor more in deep waters. The cost of drilling a well onshore (land) may range from US 2 million to 8 million.
This field specifically encounters a numerous number of problems in daily basis, so its very suitable and extremely in inquiry for advanced methodologies to mitigate those challenges and optimize the various processes

### 1.1.3 Production Operations
once the well starts producing oil and gas, production engineers and reservoir engineers continue to monitor the production trends for the well.
Throughout wells life span, the production of oil and gas from the well declines over time. Production engineers develop plans for maintaining oil flow from the well by using multiple techniques to increase or maintain the flow of oil and gas from the well. 

### 1.1.4 Abandonment
Through several years of continuous production, oil and gas production rates from the well keep declining. As mentioned earlier, engineers attempt to revive a well by well intervention and workover activities. However, after a certain point of time, it may be deemed uneconomical to revive a well given extremely poor production potential of the well. In this particular case we plug the well in respect of the environmental regulation and we call it done.


## 1.2 The Midstream Industry
The midstream industry is responsible for transporting crude oil produced from the oil wells to the downstream processing facilities. Transportation may take place via pipelines, oil vessels, trucks, or rails. In addition, the midstream industry is also responsible for the storage and marketing of petroleum commodities including, crude oil and natural gas. 

## 1.3 The Downstream Industry
The primary functions of the downstream industry include fractionation or cracking of crude oil (a process of converting thick, high molecular weight crude oil to transparent looking, low molecular weight petroleum fuels) and related chemical treatments. Which result in finished petroleum products for mass consumption as fuel.

## 2.1 Important Termenologies (AI, DM, ML, DL)?
it has been already decades. However, it may have been referred to as artificial intelligence, or by other alternate names.

I think that this would be a good time to include some definitions mainly, the difference between various terms such as AI, Data mining and Machine learning
Terminologies such as AI, ML, big data, and data mining are used interchangeably across different organizations. Therefore, it is crucial to understand the true meaning of each terminology before diving deeper into various applications. 
- AI is simply a branch of computer science that studies the simulation of human intelligence processes such as learning, reasoning and problem-solving by computers. Creating intelligent machines that work, react, and mimic cognitive functions of humans is the primary goal of AI. Examples of AI include email classification (categorization), smart personal assistants such as Siri, Alexa, and Google, automated respondents, process automation, Etc.
- Data mining is a terminology used in computer science and is defined as the process of extracting specific information from a database that was hidde. It is also called knowledge discovery in databases (KDD).
- Machine learning in the other hand is subset of AI. It is defined as the collection of using various algorithms to teach computers to find patterns in data to be used for future prediction and forecasting. ML provides computers the ability to learn without being explicitly programmed.

To set a visual difference between the two, teaching someone how to play football is ML; however, using someone to find the best football centers is data mining.

- Furthermore Deep learning is a subset of machine learning and it has basically the same functionality with an emphasis on neural networks, so basically an algorithm difference.
